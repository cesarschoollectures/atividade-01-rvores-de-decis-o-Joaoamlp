# Questão 1

Utilize o dataset Iris disponível no scikit-learn.
Divida os dados em treino e teste utilizando divisão estratificada.

**Solução**:

```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split

X, y = load_iris(return_X_y=True)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

# Questão 2

Treine um modelo utilizando `DecisionTreeClassifier`.

Depois calcule:

- acurácia no treino
- acurácia no teste

**Solução**:

```python
model = DecisionTreeClassifier()

model.fit(X_train, y_train)

accuracyTest = model.score(X_test, y_test)
print(f"AccuracyTest: {accuracyTest:.2f}")

accuracyTrain = model.score(X_train, y_train)
print(f"AccuracyTrain: {accuracyTrain:.2f}")
```

# Questão 4

Treine dez árvores com:

- max_depth = 1
- max_depth = 2
- max_depth = 3
...
- max_depth = 9
- max_depth = None

Registre em uma tabela para cada árvore:

- acurácia no treino
- acurácia no teste
- profundidade da árvore
- número de folhas

**Solução**:

```python
for depht in range(1, 10):
    model_depth = DecisionTreeClassifier(max_depth=depht)
    model_depth.fit(X_train, y_train)

    accuracyTest_depth = model_depth.score(X_test, y_test)
    print(f"AccuracyTest max_depth={depht}: {accuracyTest_depth:.2f}")

    accuracyTrain_depth = model_depth.score(X_train, y_train)
    print(f"AccuracyTrain max_depth={depht}: {accuracyTrain_depth:.2f}")

    depth_tree_depth = model_depth.get_depth()
    print(f"Depth tree max_depth={depht}: {depth_tree_depth}")

    depth_leaf_depth = model_depth.get_n_leaves()
    print(f"Depth tree max_depth={depht}: {depth_tree_depth}")

model_depth = DecisionTreeClassifier(max_depth=None)
model_depth.fit(X_train, y_train)
accuracyTest_depth = model_depth.score(X_test, y_test)
print(f"AccuracyTest max_depth=None: {accuracyTest_depth:.2f}")
accuracyTrain_depth = model_depth.score(X_train, y_train)
print(f"AccuracyTrain max_depth=None: {accuracyTrain_depth:.2f}")
depth_tree_depth = model_depth.get_depth()
print(f"Depth tree max_depth=None: {depth_tree_depth}")
depth_leaf_depth = model_depth.get_n_leaves()
print(f"Depth leaves max_depth=None: {depth_leaf_depth}")

```

**Em qual profundidade começa o overfitting?**
**Por que a árvore consegue 100% no treino quando max_depth=None?**

Na 3 podemos observar um inicio do overfitting, porem devemos levar em conta que não podemos indentificar corretamento o momento exato que ele ocorre pois não conseguimos observa o descolamento do treino em relação aos testes. 

Pois ao colocarmos none a arvore não tem limete de profuntidade assim ele vai folha a folha 


# Questão 5

Treine dois modelos:

- criterion = "gini"
- criterion = "entropy"

Compare:

- profundidade da árvore
- acurácia

**Solução**:

```python
criterions = ['gini', 'entropy']

for criterion in criterions:
    model_criterion = DecisionTreeClassifier(criterion=criterion)
    model_criterion.fit(X_train, y_train)

    accuracyTest_criterion = model_criterion.score(X_test, y_test)
    print(f"AccuracyTest criterion={criterion}: {accuracyTest_criterion:.2f}")

    accuracyTrain_criterion = model_criterion.score(X_train, y_train)
    print(f"AccuracyTrain criterion={criterion}: {accuracyTrain_criterion:.2f}")

    depth_tree_criterion = model_criterion.get_depth()
    print(f"Depth tree criterion={criterion}: {depth_tree_criterion}")
```

# Questão 6

Escolha um hiperparâmetro e investigue seu impacto.

Sugestões:

- min_samples_split
- min_samples_leaf

Mostre resultados e interprete.
- melhor modelo encontrado
- acurácia
- parâmetros

**Solução**:

```python
model_gini = DecisionTreeClassifier(criterion='gini')
model_gini.fit(X_train, y_train)
accuracyTest_gini = model_gini.score(X_test, y_test)
print(f"AccuracyTest max_depth=3 criterion=gini: {accuracyTest_gini:.2f}")
accuracyTrain_gini = model_gini.score(X_train, y_train)
print(f"AccuracyTrain max_depth=3 criterion=gini: {accuracyTrain_gini:.2f}")
depth_tree_gini = model_gini.get_depth()
print(f"Depth tree max_depth=3 criterion=gini: {depth_tree_gini}")
depth_leaf_gini = model_gini.get_n_leaves()
print(f"Depth leaves max_depth=3 criterion=gini: {depth_leaf_gini}")

model_entropy = DecisionTreeClassifier(criterion='entropy')
model_entropy.fit(X_train, y_train)
accuracyTest_entropy = model_entropy.score(X_test, y_test)
print(f"AccuracyTest max_depth=3 criterion=entropy: {accuracyTest_entropy:.2f}")
accuracyTrain_entropy = model_entropy.score(X_train, y_train)
print(f"AccuracyTrain max_depth=3 criterion=entropy: {accuracyTrain_entropy:.2f}")
depth_tree_entropy = model_entropy.get_depth()
print(f"Depth tree max_depth=3 criterion=entropy: {depth_tree_entropy}")
depth_leaf_entropy = model_entropy.get_n_leaves()
print(f"Depth leaves max_depth=3 criterion=entropy: {depth_leaf_entropy}")

```

O critério gini apresentou melhor desempenho, pois obteve maior acurácia no conjunto de teste e produziu uma árvore menos complexa, o que tende a favorecer melhor generalização no contexto de Machine Learning.
